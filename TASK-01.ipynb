```python
# Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, 
# such as the distribution of
# ages or genders in a population.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("world_population.csv")
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
df=df[["Country/Territory","2022 Population"]]
```

```python
df
```

```python
df_top10 = df.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
df_top10
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_top10["Country/Territory"], df_top10["2022 Population"], color=["skyblue","darkblue"])
plt.title("Top 10 Most Populated Countries (2022)" ,fontsize=16)
plt.xlabel("Country",fontsize=14)
plt.ylabel("Population",fontsize=18)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)  
plt.tight_layout()
plt.show()
```
